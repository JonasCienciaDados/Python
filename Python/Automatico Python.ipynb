```python
import yfinance
```

```python
ticker = input("Digite o código da ação: ")
dados = yfinance.Ticker(ticker)
```

```python
dados.history()
```

```python
tabela = dados.history("6mo")
tabela
```

```python
import pyautogui
import pyperclip
```

```python
fechamento = tabela.Close
fechamento
```

```python
fechamento.plot()
```

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]

print(maxima)
print(minima)
print(atual)
```
